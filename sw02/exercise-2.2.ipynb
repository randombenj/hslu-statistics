**teil b**

Erstellen Sie für jedes Labor einen Boxplot der Messfehler. Unter dem Messfehler eines Labors bei einer Probe verstehen wir den gemessenen Wert minus den Median über alle Labors. Welche der 21 Labors haben systematische
Fehler in ihrem Analyseverfahren? Welche haben grosse Zufallsfehler, und bei
welchen Labors ist die Qualität der Analysen besonders gut?
Python-Hinweise: Wir ziehen zunächst von jeder Spalte den Median ab

```
schlamm_centered = schlamm - schlamm.median()
```

und zeichnen den Boxplot mit vertauschten Zeilen und Spalten. Dies geschieht
mit dem Attribut .T

```
schlamm_centered.T.plot(kind="box")
```

````python
schlamm_centered = schlamm - schlamm.median()
````

````python
schlamm_centered.T.plot(kind="box")  # .T transposes columns/rows
````

- **systematische fehler:**
  
  Labor 7, 15 und 21
  

- **grosse zufallsfehler:**

  Labor 6, 12, 18, 21
  

- **gute qualitaet:**

  Labor 2, 3, 8, 10, 11, 13, 14, 17, 20
