```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
passengers = pd.read_csv("AirPassengers.csv")
passengers.head()
```

```python
passengers['TravelDate'] =  pd.to_datetime(passengers['TravelDate'])
passengers.set_index('TravelDate', inplace=True)
passengers.plot()
```
