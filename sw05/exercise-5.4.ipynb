Ein Statistiker beobachtet, dass ein Angler innerhalb von 2 Stunden 15 Fische fängt. Er nimmt an, dass es sich um einen Poissonprozess handelt und überlegt sich:

```python
import numpy as np
import scipy.stats as st

min = 2 * 60  # 12 h
fisch_mu = 15/min  # fische/minute
fisch_mu
```

```python
12 * fisch_mu
```

---
a) Mit welcher Wahrscheinlichkeit dauert es länger als 12 Minuten, bis der nächste Fisch anbeisst? *Hinweis*: Benützen Sie die Momentenmethode, um Parameter zuschätzen.

$P(X > 12)$

```python
st.expon.cdf(x=12, loc=fisch_mu)
```

b) Mit welcher Wahrscheinlichkeit beissen innerhalb der nächsten 12 Minuten genau 2 Fische an?

```python
st.expon.cdf(x=12, loc=fisch_mu*2)
```
